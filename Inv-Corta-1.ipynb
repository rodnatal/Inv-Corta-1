## a. Serie Vac√≠a

```python
# Series
import pandas as pd
import numpy as np


#1. Crear una serie vac√≠a
print('------------         Serie vac√≠a           --------------\n')
serie1 = pd.Series()
print(serie1)
```

## b.Crear una serie a partir de un arreglo (sin √≠ndice)

```python
#2. Crear una serie a partir de un arreglo sin √≠ndice
print('\n--------- Serie  desde arreglo sin √≠ndice -------------\n')
datos= np.array(['a','b','c','d','e','f'])
serie2 = pd.Series(datos)
print(serie2)

# Imprimir el primer elemento de la serie
print('El primer elemento de la serie es: ', serie2[0])
```

## c. Crear una serie a partir de un arreglo (con √≠ndice)

```python
#3. Crear una serie a partir de un arreglo con √≠ndice
print('\n--------  Serie desde arreglo con √≠ndice  ------------\n')
datos= np.array(['a','b','c','d','e','f'])
serie3 = pd.Series(datos, index=[100, 101, 102, 103, 104, 105])
print(serie3)

# Imprimir los primeros 3 elementos
print('\nLos primeros 3 elementos:\n\n',serie3[:3]) 


# Imprimir un elemento usando un √≠ndice
print('\nEl elemento con indice 102\n',serie3[102]) 
```

## d. Crear una serie a partir de un diccionario

```python
#4. Crear una serie a partir de un diccionario
print('\n-----------   Serie desde diccionario   --------------\n')
datos = {'a':0.,'b':1.,'c':2.}
serie4 = pd.Series(datos, index=['b', 'c', 'd', 'a'])
print(serie4)

# imprimir los √∫ltimos 3 elementos
print('\nLos √∫ltimos 3 elementos:\n',serie4[-3:]) 
```

## e. Crear una serie a partir de un escalar

```python
#5. Crear una serie a partir de un escalar
print('\n-----------      Serie desde escalar      -------------\n')
serie5 = pd.Series(7, index=[0, 1, 2, 3])
print(serie5)
print('\nLos primeros elementos de los √≠ndices 0, 1 y 2:\n',serie5[[0,1,2]]) 
```

## Redondeo de datos 

```python
# Redondeo de datos en dataframes
import pandas as pd

# Crear dataframe
d = {'Cuatrimestre':['Cuatri 1','Cuatri 2','Cuatri 3','Cuatri 4'],
      'Ingreso':[10000000.157, 20000000.257, 30000000.357, 40000000.457]}
dataframe=pd.DataFrame(d)


# Redondear los valores a dos decimales
print('\n-----------     Valores redondeados     ---------------\n')
pd.options.display.float_format = '{:.2f}'.format
print(dataframe)
```

```python
import pandas as pd

# Crear dataframe
d = {'Cuatrimestre':['Cuatri 1','Cuatri 2','Cuatri 3','Cuatri 4'],
      'Ingreso':[10000000.111, 20000000.222, 30000000.333, 40000000.444]}
dataframe=pd.DataFrame(d)

# Dar formato con comas y s√≠mbolo de d√≥lar
print('\n-----------   dataframe con comas   ---------------\n')
pd.options.display.float_format = '${:,.2f}'.format
print(dataframe)
```

```python
# La funci√≥n describe()
import pandas as pd
import numpy as np

#Crear un diccionario de series
d = {'Nombre':pd.Series(['Jose','Vilma','JW','Noelia','Nat','Marianela','Alina',
   'Sebasti√°n','Sof√≠a','Mat√≠as','Sara','Stella']),
   'Edad':pd.Series([69,63,35,34,33,30,28,26,21,12,8,6]),
   'Puntos':pd.Series([25,35,15,55,75,25,15,75,65,95,95,95])}

#Crear dataframe
print('\n-------------     dataframe     --------------\n')
df = pd.DataFrame(d)
print(df)


#imprimir estad√≠sticas
print('\n-------------     describe()     --------------\n')
print(df.describe())


# Calcular e imprimir la media de cada columna
print('\n-----------       mean()      -----------------\n')
print(df.mean())


# Calcular e imprimir la media de determinada columna
print('\n--------  La media para la columna Edad  ------\n')
print(df.loc[:,"Edad"].mean())


# Calcular e imprimir la mediana de cada columna
print('\n-------------     median()      ---------------\n')
print(df.median())


# Calcular e imprimir la mediana de determinada columna
print('\n-----  La mediana para la columna Edad  --------\n')
print(df.loc[:,"Edad"].median())


# Calcular e imprimir la desviaci√≥n est√°ndar
print('\n------------   desviaci√≥n est√°ndar  -------------\n')
print(df.std())


# Calcular e imprimir la desviaci√≥n est√°ndar de determinada columna
print('\n---  desviaci√≥n est√°ndar para la columna Edad ---\n')
print(df.loc[:,"Edad"].std())
```

## 1. Scale()
Se puede utilizar para estandarizar set de datos como se muestra en los siguientes ejemplos:

```python
# Ejemplo 1
# La funci√≥n scale para estandarizar un set de datos:
from sklearn import preprocessing
import numpy as np
import pandas as pd
from sklearn import datasets

linnerud=pd.DataFrame(datasets.load_linnerud().data)
linnerud_scaled=preprocessing.scale(linnerud)
print('\n----  Linnerud dataset estandarizado con scale() ------\n')
print(linnerud_scaled)
```

```python
# Ejemplo 2
# La funci√≥n scale para escalar los valores en el rango [0,1]

from sklearn import preprocessing
import numpy as np
import pandas as pd
from sklearn import datasets
linnerud=pd.DataFrame(datasets.load_linnerud().data)

# Se puede escalar un set de datos al rango dado usando MinMaxScaler

linnerud_min_max_scaler=preprocessing.MinMaxScaler()
linnerud_minmax=linnerud_min_max_scaler.fit_transform(linnerud)
print('El valor m√≠nimo despu√©s de escalar es: ', linnerud_minmax.min())
print('El valor m√°ximo despu√©s de escalar es: ', linnerud_minmax.max())
print('\n----  Linnerud dataset estandarizado al rango [0,1]------\n')
linnerud_minmax
```

## 2. Transformaciones no lineales
 Hay dos tipos de transformaciones: "Quantile" y "Power transformations".
 
 "Quantile transformation" mapea datos a una distribuci√≥n uniforme mientras que  "Power transformations" son una familia de transformaciones param√©tricas que  ayudan a mapear datos de la manera m√°s parecida a una distribuci√≥n gaussiana.


```python
# 2. Transformaciones no lineales
# Existen dos tipos de transformaciones: "Quantile Transformations" y 
# "Power transformations"

# Ejemplo: Mapear a una distribuci√≥n uniforme:
from sklearn import preprocessing
import pandas as pd
import numpy as np
from sklearn.datasets import load_linnerud
from sklearn.model_selection import train_test_split

linnerud=load_linnerud()
X,y=linnerud.data, linnerud.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
quantile_trans = preprocessing.QuantileTransformer(random_state=0)
X_train_trans = quantile_trans.fit_transform(X_train)
X_test_trans = quantile_trans.transform(X_test)
np.percentile(X_train[:, 0], [0,25,50,75,100])
np.percentile(X_test[:, 0], [0,25,50,75,100])
np.percentile(X_test_trans[:, 0], [0,25,50,75,100])
```

## 3. Normalizaci√≥n

Es el proceso de escalar muestras individuales para tener normas de 1

## 4. Codificaci√≥n
Se utiliza cuando algunas caracter√≠sticas de los datos son dados como categor√≠as, por ejemplo: [hombre, mujer]. Dichas categor√≠as pueden ser convertidas a enteros con el fin de que la manipulaci√≥n de los datos sea m√°s sencilla.

```python
# 4. Codificaci√≥n de categor√≠as

from sklearn import preprocessing
codificador = preprocessing.OrdinalEncoder()

X = [['hombre','adulto','Heredia'],['mujer','menor','Cartago']]
codificador.fit(X)
codificador.transform([['mujer', 'adulto', 'Cartago']])
```

## 5. Discretizaci√≥n
Se utiliza para convertir datos con caracter√≠sticas en valores discretos. La funci√≥n Binarizer convierte valores num√©ricos a valores booleanos


```python
# 5. Discretizaci√≥n
from sklearn import preprocessing
datos = [[-1, 0, 1],
         [2, -1, 0],
         [1, 2, -1]]
binarizer = preprocessing.Binarizer().fit(datos)
binarizer
binarizer.transform(datos)
```

## 6. Impute
Utilizado cuando hay datos faltantes o no v√°lidos

```python
# 6. Una alternativa para cuando faltan datos es el modulo impute 
# de sklearn
from sklearn import preprocessing
import numpy as np
from sklearn.impute import SimpleImputer


datos = [[-1, 0, 1],
         [2, np.nan, 0],
         [1, 2, -1]]

imputer= SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(datos)
print(imputer.transform(datos))
```
